# Get innovation examples for high/low covariates
In which we get good example innovations at the high and low end of the covariate scale.

```python
import pandas as pd
import sys
import data_handler
reload(data_handler)
from data_handler import get_growth_decline_words_and_params, get_growth_words, get_all_covariates, melt_frames
```

## Load data

```python
failure_words, _ = get_growth_decline_words_and_params()
success_words = get_growth_words()
innovations = failure_words + success_words
covariates, covariate_names = get_all_covariates()
covariates = map(lambda c: c.loc[innovations, :], covariates)
combined_data = melt_frames(covariates, covariate_names)
print(combined_data.head())
# compute mean values
combined_data_mean = combined_data.drop('date', axis=1, inplace=False).groupby('word').apply(pd.np.mean)
print(combined_data_mean.head())
```

```python
example_covariates = ['C3', 'DU', 'DS', 'DT']
# restrict to high frequency to start
lower_pct = 90
f_lower = pd.np.percentile(combined_data_mean.loc[:, 'f'], lower_pct)
combined_data_cutoff = combined_data_mean[combined_data_mean.loc[:, 'f'] >= f_lower]
for c in example_covariates:
    print('high/low values for covariate %s'%(c))
    cov_vals = combined_data_cutoff.loc[:, c].sort_values(inplace=False, ascending=False)
    print(cov_vals)
```
